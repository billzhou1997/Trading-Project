```python
from alpaca_trade_api.rest import REST, TimeFrame
# api = tradeapi.REST(key_id= PUB_KEY, secret_key=SEC_KEY, base_url=BASE_URL)
symb = "SPY"                    
data=api.get_bars(symb, TimeFrame.Minute, "2022-07-25",  "2022-07-29", limit=10, adjustment='raw')
#?api.getbars
#print(type(data[0]))
#print((data[0]))

data_dict = data._raw
print(type(data_dict))
data_frame={}
for symb in symb_list:
    data_frame[symb]=[]
    data=api.get_bars(symb, TimeFrame.Day, "2022-07-24",  "2022-07-28", limit=5, adjustment='raw')
    data_dict = data._raw
    for i in range(4):
        data_frame[symb].append(data_dict[i]["c"])

change={}
for item in data_frame:
    dif=(data_frame[item][-1]-data_frame[item][0])
    avg=sum(data_frame[item])/len(data_frame[item])
    change[item]=dif/avg
print(change)

sorted_values = sorted(change.values())
sorted_dict = {}

for i in sorted_values:
    for k in change.keys():
        if change[k] == i:
            sorted_dict[k] = change[k]
            break
print(sorted_dict)
lst1=[]
lst2=[]
for i,key in enumerate(sorted_dict):
    if i<2:
        lst1.append(key)
    if i>3:
        lst2.append(key)
print(lst1,lst2)
```

```python
import alpaca_trade_api as tradeapi
import numpy as np
import time
SEC_KEY = 'G0Kecjo7cETmxzImpRURMWhIPeic9SsBNrmsTajl'
PUB_KEY = 'PKS4CGU2409PM5WOJ7KM'
BASE_URL = 'https://paper-api.alpaca.markets'
api = tradeapi.REST(key_id= PUB_KEY, secret_key=SEC_KEY, base_url=BASE_URL)

symb = "SPY"

while True:
    print("")
    print("Checking Price")
    
    market_data = api.get_bars(symb, 'minute', limit=5) # Get one bar object for each of the past 5 minutes
    api.get_bars(symb)
    close_list = [] # This array will store all the closing prices from the last 5 minutes
    for bar in market_data[symb]:
        close_list.append(bar.c) # bar.c is the closing price of that bar's time interval

    close_list = np.array(close_list, dtype=np.float64) # Convert to numpy array
    ma = np.mean(close_list)
    last_price = close_list[4] # Most recent closing price

    print("Moving Average: " + str(ma))
    print("Last Price: " + str(last_price))
    
    time.sleep(60) # Wait one minute before retreiving more market data
```
